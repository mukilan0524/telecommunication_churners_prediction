```python
copy_dataset=data_set.copy()
```

```python
copy_dataset.head()
```

```python
copy_dataset.info()
```

```python
copy_dataset["TotalCharges"]=pd.to_numeric(copy_dataset["TotalCharges"],errors="coerce")
```

```python
copy_dataset.info()
```

```python
copy_dataset.isnull().sum()
```

```python
copy_dataset.columns.values
```

```python
copy_dataset.loc[copy_dataset['TotalCharges'].isnull()]
```

```python
copy_dataset.dropna(how = 'any', inplace = True)
copy_dataset.isnull().sum()
```

```python
copy_dataset.head()
```

```python
copy_dataset.drop("customerID",axis=1,inplace=True)
copy_dataset.info()
```

```python
copy_dataset["Churn"]=np.where(copy_dataset["Churn"]=="Yes",1,0)
```

```python
dummy_set=pd.get_dummies(copy_dataset)
```

```python
dummy_set.head()
```

```python
plt.figure(figsize=(20,8))
corr_matrix=dummy_set.corr()
sns.heatmap(corr_matrix)
plt.show()
```

```python
plt.figure(figsize=(20,8))
corr_matrix["Churn"].sort_values(ascending=False).plot(kind="bar",)
```

```python
dummy_set.to_csv("Churn_analysis_dataset.csv")
```
